## 2. Image Tensor Manipulation with numpy 

In this task, you will work with image data represented as NumPy arrays. You will perform basic image manipulation operations using NumPy, such as flipping, rotating, and resizing images. This exercise will help you understand how to manipulate image data using NumPy arrays and visualize the results.

**Objective:**
- To perform basic image manipulation operations using NumPy arrays.

**Requirements:**

1. **Load an Image**: Load an image from your local file system using Pillow and convert it to a NumPy array.
2. **Rotate the Image**: Rotate the image 90 degrees clockwise
3. **Convert to Grayscale**: Convert the color image to grayscale.
4. **Increase Image Brightness**: Increase the brightness of the image by 10%.
5. **Decrease Image Brightness**: Decrease the brightness of the image by 10%.
6. **Increase Image Contrast**: Increase the contrast of the image by 10%.
7. **Decrease Image Contrast**: Decrease the contrast of the image by 10%.

**Expected Output:**
- The script should sequentially perform the mentioned operations and display the manipulated image at the end of each task.


````python
# Open an image file and convert to numpy array
import numpy as np
import matplotlib.pyplot as plt

img = Image.open("image.jpg")
np_img = np.array(img)

plt.imshow(np_img, cmap='gray' if len(np_img.shape) == 2 else None)
plt.title("Original Image")
plt.axis('off')
plt.show()
````

````python
# Rotate the image 90 degrees clockwise
rotated_img = np.rot90(np_img, k=-1)  # k=-1 for clockwise, 1 for antickock wise

plt.imshow(rotated_img, cmap='gray' if len(np_img.shape) == 2 else None)
plt.title("Rotated 90 Degrees Clockwise")
plt.axis('off')
plt.show()

# Convert to grayscale
gray_img = img.convert('L').convert('RGB') #note, without converting to RBG first, passing to np.array is kaputt
numpy_gray = np.array(gray_img)

plt.imshow(numpy_gray, cmap='gray' if len(np_img.shape) == 2 else None)
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
````

````python
# Increase the brightness by 10%
bright_enhancer = ImageEnhance.Brightness(img)
bright_img = bright_enhancer.enhance(1.5)
bright_img = np.array(bright_img)

plt.imshow(bright_img, cmap='gray' if len(np_img.shape) == 2 else None)
plt.title("Brightness Increased by 10%")
plt.axis('off')
plt.show()

# Decrease the brightness by 10%
dim_img = bright_enhancer.enhance(0.5)
dim_img = np.array(dim_img)

plt.imshow(dim_img, cmap='gray' if len(np_img.shape) == 2 else None)
plt.title("Brightness Decreased by 10%")
plt.axis('off')
plt.show()
````

````python
# Increase the contrast by 10%
contrast_enhancer = ImageEnhance.Contrast(img)
high_contrast_img = contrast_enhancer.enhance(1.5)

plt.imshow(high_contrast_img, cmap='gray' if len(np_img.shape) == 2 else None)
plt.title("Contrast Increased by 10%")
plt.axis('off')
plt.show()

# Decrease the contrast by 10%
low_contrast_img = contrast_enhancer.enhance(0.5)

plt.imshow(low_contrast_img, cmap='gray' if len(np_img.shape) == 2 else None)
plt.title("Contrast Decreased by 10%")
plt.axis('off')
plt.show()
````
